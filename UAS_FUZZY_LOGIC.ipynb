# Metode Tsukamoto

```python

# Fungsi keanggotaan suhu
def mu_suhu_normal(x):
    if 36 <= x <= 37.5:
        return (37.5 - x) / (37.5 - 36)
    return 0

def mu_suhu_demam(x):
    if 37 <= x <= 39:
        return (x - 37) / (39 - 37)
    return 0

def mu_suhu_tinggi(x):
    if x >= 38:
        return (x - 38) / (41 - 38)
    return 0


# Fungsi keanggotaan trombosit
def mu_trombosit_rendah(x):
    return 1 if x <= 150 else 0

def mu_trombosit_normal(x):
    return 1 if x > 150 else 0


# Fungsi keanggotaan leukosit
def mu_leukosit_rendah(x):
    return 1 if x <= 4 else 0

def mu_leukosit_normal(x):
    return 1 if x > 4 else 0


# Data pasien
suhu = 38.5
trombosit = 120
leukosit = 3.8

# Derajat keanggotaan
a1 = min(mu_suhu_demam(suhu), mu_trombosit_rendah(trombosit))
a2 = min(mu_suhu_tinggi(suhu), mu_leukosit_normal(leukosit))

# Nilai z (konsekuen Tsukamoto)
z1 = 90   # Risiko Tinggi
z2 = 66.67  # Risiko Sedang

# Defuzzifikasi
z_final = (a1 * z1 + a2 * z2) / (a1 + a2)

print("Nilai Risiko DBD =", round(z_final, 2))

if z_final >= 80:
    print("Diagnosa: Risiko Tinggi")
elif z_final >= 60:
    print("Diagnosa: Risiko Sedang")
else:
    print("Diagnosa: Risiko Rendah")
```

# ANFIS

```python

# Input
x1 = 60  # soil moisture
x2 = 26  # avg temperature

# Parameter tiap rule (p, q, r)
rules = [
    (0.0692, 0.0461, 0.00231),
    (0.0698, 0.0390, 0.00139),
    (0.0585, 0.0183, 0.00084),
    (0.0505, 0.0318, 0.00115)
]

# Hitung output tiap rule
outputs = []
for p, q, r in rules:
    f = p * x1 + q * x2 + r
    outputs.append(f)

# Defuzzifikasi (rata-rata)
hasil = sum(outputs) / len(outputs)

print("Prediksi Produksi Jagung =", round(hasil, 4), "ton/ha")

```
